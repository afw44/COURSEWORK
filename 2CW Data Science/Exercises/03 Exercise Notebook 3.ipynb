## 3b) Fit a linear model $f(x) = \beta_0 + \beta_1 x$ to the data

- Print the values of the regression coefficients $\beta_0, \beta_1$ determined using least-squares. 
- Plot the fitted model and the scatterplot on the same plot.
- Compute and print the **MSE** and the $R^2$ coefficient for the fitted model.

All numerical outputs should be displayed to three decimal places.

```python



def polyreg(data_matrix, k):
        
    xvals = data_matrix[:,0]
    yvals = data_matrix[:,1]
    
    Xk = np.column_stack(tuple([np.ones(xvals.shape)] + 
                               [xvals**i for i in range(1,k+1)]))
        
    beta_vec = np.linalg.lstsq(Xk, yvals)[0]
    fit = Xk.dot(beta_vec)
    residuals = yvals - fit
    
    return beta_vec, fit, residuals, Xk


fig1, ax = plt.subplots(1,1,figsize=(4,4))





PR1 = polyreg(np.column_stack((xvals,yvals)), 1)
linear_betas = PR1[0]
linear_fit = PR1[3].dot(PR1[0])
linear_sse = np.sum((yvals - linear_fit)**2)
linear_mse = linear_sse / np.shape(yvals)[0]

PR0 = polyreg(np.column_stack((xvals,yvals)), 0)
constant_ss =  np.sum((yvals - PR0[3].dot(PR0[0]))**2)

R2 = 1 - linear_sse / constant_ss

ax.scatter(xvals, yvals,
           color='black', marker='x', s=15,
           label='Data')

ax.plot(xvals, linear_fit,
        color='red', linewidth=3, 
        label=f'Linear fit:'
              f'\n{round(linear_betas[0],-1)} + {round(linear_betas[1],2)}t'
              f'\nR^2 = {round(R2,3)}')

ax.legend(loc='best')
ax.set_xlabel('Counting period')
ax.set_ylabel('Tube Total (000s)')
ax.set_title(f'Tube ridership over time\nMSE={round(linear_mse,2)}')



residuals = PR1[2]
fig2 = plt.figure(figsize=(4,8))
spec = fig2.add_gridspec(4,1)

ax0 = fig2.add_subplot(spec[0:2,0]) 
ax1 = fig2.add_subplot(spec[2,0])
ax2 = fig2.add_subplot(spec[3,0])



ax0.scatter(xvals, residuals)





```
